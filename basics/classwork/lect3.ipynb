# **Exercises**

```python
#calculate area of rectangle

#user input
a = int(input("Enter the length of rectangle: "))
b = int(input("Enter the breadth of the rectangle: "))

#logic block
area = a*b

#output function
print("The area of rectangle is: ", area)
```

```python
#calc simple interest

#user input
principal = int(input("Enter the amount: "))
rate = 7.5 #given
time = float(input("Enter the time period: "))

#logic block
simple_int = (principal*rate*time)/100

#output
print("The simple interest is:", simple_int, "for a period of", time, "years.")
```

```python
#validate age using if...else

#user input
a = int(input("Enter the age: "))

#logic block
if a<=18:
  print("Invalid age! You are not eligible for the program.")
else:
  print("Congratulations! You are eligible for the program, hurray!")
```

# **FOR loop variation**

```python
#for direct (output starts from 0)
for i in range(10):
  print(i)
```

```python
#for with start and end value
for i in range(1,10+1):
  print(i)
```

```python
#for with start, end, step
for i in range(1,10+1,3):
  print(i)
```

```python
#for with user entered start-end
start = int(input("Enter the start number: "))
end = int(input("Enter the end number: "))

for i in range(start, end+1):
  print(i)
```

```python
#print even numbers in user given range
start = int(input("Enter the start number: "))
end = int(input("Enter the end number: "))

for i in range(start,end+1):
  if(i%2==0):
    print(i)
```

```python
#generate tables between 1 and 20
for i in range(1,20+1):
  print()
  for j in range(1,10+1):
    print(i,"x",j, "=", i*j)
```

FOR vs WHILE

```python
#table of the number taken by user input
a = int(input("Enter the number: "))
i = 1

while(i<=10):
  ans = a*i
  print(a,"x", i,"=",ans)
  i=i+1
```
