```python
#palindrome or not
num = int(input("Enter the number to be reversed: "))
rev = 0
numm=num
#because num changes through logic block, original num val should be stored somewhere

#code for reversing the string to check if palindrome or not
while(num>0):
  rem=num%10
  rev=rev*10+rem
  num//=10 #num=num/10

#code for checking if palindrome or not
if(rev==numm):
  print("The string is a palindrome.")
else:
  print("The string is not a palindrome.")
```

```python
#factorial using for loop
n = int(input("Enter number: "))
fact = 1
for i in range(1,n+1):
  fact = fact*i
print(fact)
```

```python
#sum of first 'n' natural numbers using for loop
num = int(input("How many natural numbers do you want? "))

sum=0
for i in range(1,num+1):
  sum=sum+i
print(sum)

#using while loop and user input
num = int(input("How many natural numbers do you want? "))
j=1
sum = 0
while(j<=num):
  sum=sum+j
  j+=1
print(sum)
```
